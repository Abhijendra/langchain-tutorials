### 3. Custom Tools

```python
from langchain.tools import tool
```

```python
# step 1: create a function

def multiply(a, b):
    '''multiply a and b'''
    return a * b 
```

```python
# step 2: Add type hint

def multiply(a:int, b:int) -> int:
    '''multiply a and b'''
    return a * b
```

```python
# step 3: add tool decorator

@tool
def multiply(a:int, b:int) -> int:
    '''multiply a and b'''
    return a * b
```

```python
result = multiply.invoke({'a':5,'b':3})
```

```python
result
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())
```
