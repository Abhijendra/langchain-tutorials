```python
from langchain.tools import tool
```

```python
@tool
def add(a:int, b:int) -> int:
    '''adds two numbers'''
    return a + b 

@tool
def multiply(a:int, b:int) -> int:
    '''multiply two numbers'''
    return a * b 
```

```python
class MathToolKit:

    def get_tools(self):
        return [add, multiply]
```

```python
mathtoolkit = MathToolKit()

tools = mathtoolkit.get_tools()

for tool in tools:
    print(tool.name, '-->', tool.description)
    
```
