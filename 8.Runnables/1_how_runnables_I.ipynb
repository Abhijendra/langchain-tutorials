```python
import random
```

## Creating LLM

```python
class NakliLLM:

    def __init__(self):
        print('LLM created')

    def predict(self, prompt):

        response_list = ['AI stand for Arti Intelligence.',
                        'Delhi is the capital of India.',
                        'Google is a large US based company.']

        return {'response': random.choice(response_list)}
```

```python
llm = NakliLLM()
```

```python
llm.predict('Capital of India?')
```

## Creating Prompt Template

```python
class PromptTemplate:
    
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
prompt = PromptTemplate(template='Write a {length} paragraph on {topic}',
                          input_variables=['length','topic'])

query_dict = {'length':'long','topic':'G7 countries'}

print(prompt.format(input_dict=query_dict))
```

```python
llm.predict(prompt=prompt)
```

## Creating Chain

```python
class NakliLLMChain:

    def __init__(self, llm, prompt):
        self.llm = llm 
        self.prompt = prompt
        print('Chain created')

    def run(self, input_dict):

        prompt = self.prompt.format(input_dict)
        result = self.llm.predict(prompt)
        return result['response']
```

```python
llm = NakliLLM()
```

```python
prompt = PromptTemplate(template='Write a {length} paragraph on {topic}',
                          input_variables=['length','topic'])

# print(prompt.format(input_dict=query_dict))

chain = NakliLLMChain(llm=llm, prompt=prompt)
```

```python
query_dict = {'length':'medium length','topic':'China Moon Landing'}
chain.run(query_dict)
```
