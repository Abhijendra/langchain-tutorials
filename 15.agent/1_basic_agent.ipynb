```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool 
import requests

from langchain_community.tools import DuckDuckGoSearchRun
```

```python
ddgs = DuckDuckGoSearchRun()
result = ddgs.invoke('Top news in India Today')
result
```

```python
llm = ChatOpenAI(model='gpt-3.5-turbo-0125')
```

```python
from langchain.agents import create_react_agent, AgentExecutor
from langchain import hub
```

```python
# Pull ReAct agent prompt from Langchain hub

prompt = hub.pull('hwchase17/react') 
```

```python
prompt.template
```

```python
# create react agent

agent = create_react_agent(
    llm=llm,
    tools=[ddgs],
    prompt=prompt
)
```

```python
# wrap it with agentExecutor(agent will command the agent executor for tasks like API call, calculation, etc)

agent_executor = AgentExecutor(
    agent=agent,
    tools=[ddgs],
    verbose=True
)
```

```python
agent_executor.invoke({'input':'When and where Donald Trump will meet President Putin'})
```
